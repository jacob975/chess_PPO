```python
from stable_baselines3 import ppo
import gym
import time
import numpy as np
env = gym.make('BipedalWalker-v3')
```

```python
# Create a PPO agent
agent = ppo.PPO('MlpPolicy', env, verbose=1)
```

```python
# Train on the environment with 1M timesteps and save the best model
agent.learn(total_timesteps=1000000, log_interval=10, tb_log_name="ppo_bipedalwalker")
```

```python
# Load the trained agent
agent = ppo.PPO.load("ppo_bipedalwalker")
```

```python
# Evaluate the agent with animation
obs = env.reset()

for i in range(1000):
    action, _states = agent.predict(obs, deterministic=False)
    obs, rewards, dones, info = env.step(action)
    if dones:
        obs = env.reset()
        break
    env.render()
    time.sleep(0.01)
    print(i)
```

```python
env.close()
```
